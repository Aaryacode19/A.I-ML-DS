<h2>Lec - 92 | Plotting From Pandas DataFrame 2</h2>

```python
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]','', regex=True) 
# Regular expression (regex) is used to define a pattern for finding specific characters
# Here, we use regex to match $, commas, and dots and remove them from the 'Price' column
# regex=True tells Pandas to treat the pattern as a regular expression
car_sales
```

```python
type(car_sales["Price"][0])
```

```python
#Removing last 2 zeroes & use .str since price col is of string dtype
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
#adding new column
car_sales["Sale Date"] = pd.date_range("1/1/2024", periods=len(car_sales))
# pd.date_range generates a range of dates starting from 1st Jan 2024
# The number of dates equals the number of rows in the 'car_sales' DataFrame
car_sales
```

```python
#making total sales col
car_sales["Total Sales"] = car_sales["Price"].cumsum() 
#Here we added new col names total sales inside which we have stored the total of price column
car_sales
```

<p>We have got output in such a way since our price column is of string format/ datatype, so this cumsum just concanated the strings and stored in new column, refer next cell to know data type of price col</p>

```python
type(car_sales["Price"][0])
```

<p>Therefore now we will first convert the price column into integers using the astype(int) method and then call the cumsum method</p>

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum() 
car_sales
```

```python
#Lets plot the total sales, how it looks like over time
car_sales.plot(x="Sale Date", y="Total Sales");
```

```python
#reassign price col to int, since when we did it earlier it was temporary we need to reassign it again
car_sales["Price"] = car_sales["Price"].astype(int)
#Plot scatter plot with price col as numeric
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter")
```

<h2>Lec - 96 | Plotting From Pandas DataFrame 6</h2>

```python
# Import necessary library
#OO Method From Scratch
import matplotlib.pyplot as plt

# OO Method from scratch: Create a figure and axes
fig, ax = plt.subplots(figsize=(10, 6))  # Create a figure (fig) and axes (ax) with specified size

# Plot the data: Scatter plot of age (x-axis) vs cholesterol levels (y-axis)
scatter = ax.scatter(x=over_50["age"],      # x-axis data: Age column from the DataFrame
                     y=over_50["chol"],     # y-axis data: Cholesterol column from the DataFrame
                     c=over_50["target"])   # Color the points based on the target column (e.g., 0 or 1)

# Customize the plot
ax.set(title="Heart Disease and Cholesterol levels",  # Set the title of the plot
       xlabel="Age",                                 # Label for the x-axis
       ylabel="Cholesterol")                        # Label for the y-axis

# Add a legend to explain the color coding of the dots
ax.legend(*scatter.legend_elements(),     # Generate legend elements automatically from the scatter plot
          title="Target")                 # Set the title of the legend to "Target"

# Add a horizontal line to indicate the mean cholesterol level
ax.axhline(over_50["chol"].mean(),        # Calculate and place the line at the mean cholesterol value
           linestyle="--");                # Use a dashed line style for the horizontal line

# The final result: A scatter plot showing age vs. cholesterol levels,
# colour-coded by the "target" column, with a legend and a mean cholesterol line.
```

```python
#Subplot of chol, age  ,thalach
fog, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10))

#add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

#customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        xlabel="Age",
        ylabel="Cholesterol")

#add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

#add a mainline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");

#add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])

#customize ax1
ax1.set(title="Heart Disease and Thalach Levels(i.e. max heart rate)",
        xlabel="Age",
        ylabel="Thalach")

#add a legend to ax1
ax1.legend(*scatter.legend_elements(),
           title="Target")

#add a mainline to ax1
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");
```

<p>You can also use <b>sharex=True</b> to keep a common x label nd x axis for both subplots also remove xlabel of ax0</p>

```python
#Subplot of chol, age  ,thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                               sharex=True)

#add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

#customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")

#add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

#add a mainline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");

#add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])

#customize ax1
ax1.set(title="Heart Disease and Thalach Levels(i.e. max heart rate)",
        xlabel="Age",
        ylabel="Thalach")

#add a legend to ax1
ax1.legend(*scatter.legend_elements(),
           title="Target")

#add a mainline to ax1
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");

#add a title to figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

<h2>Lec - 99 | Customizing Plots 2</h2>

```python
#reset the style/ set the style
plt.style.use("seaborn-v0_8-whitegrid")

# OO Method from scratch: Create a figure and axes
fig, ax = plt.subplots(figsize=(10, 6))  

# Plot the data: Scatter plot of age (x-axis) vs cholesterol levels (y-axis)
scatter = ax.scatter(x=over_50["age"],      
                     y=over_50["chol"],     
                     c=over_50["target"],
                     cmap="winter");  #set cmap="winter" will change the colour theme of ur graph, u can use winter, summer etc refer doc

# Customize the plot
ax.set(title="Heart Disease and Cholesterol levels",  
       xlabel="Age",                                 
       ylabel="Cholesterol")                        

# Add a legend to explain the color coding of the dots
ax.legend(*scatter.legend_elements(),     
          title="Target")                 

# Add a horizontal line to indicate the mean cholesterol level
ax.axhline(over_50["chol"].mean(),        
           linestyle="--");  
```

```python

#Subplot of chol, age  ,thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                               sharex=True)

#add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter");

#customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")
#change the x axis limits
ax0.set_xlim([50, 80])

#add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

#add a mainline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");

#add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="summer");

#customize ax1
ax1.set(title="Heart Disease and Thalach Levels(i.e. max heart rate)",
        xlabel="Age",
        ylabel="Thalach")

#change x axis nd y axis limits
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

#add a legend to ax1
ax1.legend(*scatter.legend_elements(),
           title="Target")

#add a mainline to ax1
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");

#add a title to figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
